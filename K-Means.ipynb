Data processing

```python
#Data reading
import numpy as np
import pandas as pd
DATA = pd.read_csv("kmeans.csv", delimiter=',', index_col='Object')
```

```python
#Removing a Cluster column from a dataset
coords = DATA.drop('Cluster', axis=1)
```

```python
#Connecting the KMeans model from sklearn library
from sklearn.cluster import KMeans
```

```python
#Initialization with centroids
#KMeans(n_clusters=3, init=np.array([[12.6, 8.8], [13.0, 12.6], [13.0, 11.8]]), max_iter=100, n_init=1)
centroid = np.array([[11.6, 12.4], [12.13, 9.13], [1.5, 18.0]])
kmeans =KMeans(n_clusters=3, init=np.array([[12.6, 8.8], [13.0, 12.6], [13.0, 11.8]]), max_iter=100, n_init=1)
```

```python
#Model training
model = kmeans.fit(coords)
```

```python
#Assigned clusters
model.labels_.tolist()
```

```python
#Calculation of distances to centroids
alldistances = kmeans.fit_transform(coords)
```

```python
#Distances from the centroids to all data points
alldistances
```

```python
indices=[0,1,5,6,7]
DISTANCES_0 = [alldistances[i][0] for i in indices] 
```
